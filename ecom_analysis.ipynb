```python
unavailable_orders_per_day.rename(columns = {'order_purchase_day' : 'date'}, inplace = True)
```

```python
unavailable_orders = dt.merge(unavailable_orders_per_day, how = 'left', on = 'date')
unavailable_orders
```

```python
unavailable_orders['lag_unavailable_orders'] = unavailable_orders['unavailable_orders'].shift(1)
unavailable_orders
```

```python
#   nan  unavailable_orders  lag_unavailable_orders  1, 
#  ,         
unavailable_orders.isna().sum(axis=1) == 1
```

```python
unavailable_orders['change'] = pd.Series((unavailable_orders.isna().sum(axis=1) == 1)*1).values
unavailable_orders
```

```python
np.logical_or(unavailable_orders['unavailable_orders'].isna(),
              np.logical_and(unavailable_orders['unavailable_orders'].isna(),
                             unavailable_orders['lag_unavailable_orders'].isna()))
```

```python
unavailable_orders['incorrect_session'] = pd.Series(
    np.logical_or(unavailable_orders['unavailable_orders'].isna(),
              np.logical_and(unavailable_orders['unavailable_orders'].isna(),
                             unavailable_orders['lag_unavailable_orders'].isna()))*1).values
```

```python
#        unavailable_orders_per_day,
#     
unavailable_orders = unavailable_orders[unavailable_orders['incorrect_session']==0].copy()
unavailable_orders
```

```python
unavailable_orders['session_number'] = unavailable_orders['change'].cumsum()
unavailable_orders
```

```python
unavailable_orders_sessions = unavailable_orders.groupby('session_number')[['date', 'unavailable_orders']].\
        agg({'date': 'count', 'unavailable_orders': 'sum'}).reset_index().\
        rename(columns = {'unavailable_orders' : 'total_unavailable_orders',
                         'date' : 'days_duration'}).\
        sort_values(['total_unavailable_orders', 'days_duration'], ascending=[False, False])        
unavailable_orders_sessions
```

```python
#  2   
# -  5  12
# -  58  65
#     
unavailable_orders_sessions.sort_values('session_number', ascending = True).\
            set_index('session_number')['total_unavailable_orders'].plot()
```
