```python
unavailable_orders_per_day.rename(columns = {'order_purchase_day' : 'date'}, inplace = True)
```

```python
unavailable_orders = dt.merge(unavailable_orders_per_day, how = 'left', on = 'date')
unavailable_orders
```

```python
unavailable_orders['lag_unavailable_orders'] = unavailable_orders['unavailable_orders'].shift(1)
unavailable_orders
```

```python
# Если сумма nan значений unavailable_orders и lag_unavailable_orders равна 1, значит
# смениля период, когда присутствовали данные ошибки и они исчезли или наоборот
unavailable_orders.isna().sum(axis=1) == 1
```

```python
unavailable_orders['change'] = pd.Series((unavailable_orders.isna().sum(axis=1) == 1)*1).values
unavailable_orders
```

```python
np.logical_or(unavailable_orders['unavailable_orders'].isna(),
              np.logical_and(unavailable_orders['unavailable_orders'].isna(),
                             unavailable_orders['lag_unavailable_orders'].isna()))
```

```python
unavailable_orders['incorrect_session'] = pd.Series(
    np.logical_or(unavailable_orders['unavailable_orders'].isna(),
              np.logical_and(unavailable_orders['unavailable_orders'].isna(),
                             unavailable_orders['lag_unavailable_orders'].isna()))*1).values
```

```python
# количество строк совпадает с изначальным в группировке unavailable_orders_per_day,
# значит периоды  определены верно
unavailable_orders = unavailable_orders[unavailable_orders['incorrect_session']==0].copy()
unavailable_orders
```

```python
unavailable_orders['session_number'] = unavailable_orders['change'].cumsum()
unavailable_orders
```

```python
unavailable_orders_sessions = unavailable_orders.groupby('session_number')[['date', 'unavailable_orders']].\
        agg({'date': 'count', 'unavailable_orders': 'sum'}).reset_index().\
        rename(columns = {'unavailable_orders' : 'total_unavailable_orders',
                         'date' : 'days_duration'}).\
        sort_values(['total_unavailable_orders', 'days_duration'], ascending=[False, False])        
unavailable_orders_sessions
```

```python
# видим 2 ярковыраженных скопления 
# - между 5 и 12
# - между 58 и 65
# посмотрим на эти данные поближе
unavailable_orders_sessions.sort_values('session_number', ascending = True).\
            set_index('session_number')['total_unavailable_orders'].plot()
```

```python
# Разберемся с заказами, сгруппируем товары по заказу
items_per_order = order_items.groupby('order_id')['product_id'].apply(list).reset_index().\
                rename(columns = {'product_id' : 'items_list'})
items_per_order
```

```python
# добавим количество товаров в заказе и количество уникальных товаров
items_per_order['order_length'] = items_per_order['items_list'].apply(lambda x: len(x))
items_per_order['unique_items'] = items_per_order['items_list'].apply(lambda x: len(set(x)))
items_per_order
```

```python
orders_main = items_per_order.merge(order_items.groupby('order_id')['price'].sum().reset_index(),
                                    how = 'left', on = 'order_id')
orders_main
```

```python
orders_main = orders.merge(orders_main, how = 'left', on = 'order_id')
orders_main
```

```python
orders_main['month'] = orders_main['order_purchase_timestamp'].to_numpy().astype('datetime64[M]')
orders_main['day'] = orders_main['order_purchase_timestamp'].to_numpy().astype('datetime64[D]')
```

```python
orders_main.groupby(['month', 'order_status'])['order_id'].count()
```

```python
orders_status = orders_main.groupby(['month', 'order_status'])['order_id'].count().to_frame()\
                .transpose().stack(0).reset_index()
orders_status
```

```python
orders_status.drop('level_0', axis = 1, inplace = True)
orders_status = orders_status.rename_axis(None, axis=1)
orders_status
```

```python
orders_status = orders_status.fillna(0)
orders_status
```

```python
orders_status = orders_status.set_index('month')
orders_status.sum(axis = 1).values
```

```python
# покупка последнего заказа была осуществлена 08-29, а дата последней доставки заказа 10-17
# то есть последние два месяца  и первые 3 при анализе можно не учитывать
orders_status_pct = orders_status.divide(orders_status.sum(axis = 1).values, axis = 0).round(3)*100
orders_status_pct = orders_status_pct[np.logical_and(orders_status_pct.index >= '2017-01-01',
                                             orders_status_pct.index <= '2018-08-01')].copy()
orders_status_pct
```

```python
orders_status = orders_status[np.logical_and(orders_status.index >= '2017-01-01',
                                             orders_status.index <= '2018-08-01')].copy()
```
